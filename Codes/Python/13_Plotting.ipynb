# Case Study: Time taken by fib(n)

```python
def fibslow(n):
    if n == 0 or n == 1:
        return n
    else:
        return fibslow(n-1) + fibslow(n-2) 
```

```python
fibslow(4)
```

```python
def fibfast(n):
    if n<=1:
        return n
    a , b = 0 , 1
    for i in range(a , b):
        a,b = b , a+b
    return a
```

```python
fibfast(4)
```

```python
from_n , to_n = 1 , 30

x = list(range(from_n, to_n))

print(x)
```

```python
def time_function(fn , start , end):
    from datetime import datetime
    
    times = []
    
    for i in range (start , end):
        start_time = datetime.now()
        _ = fn(i)
        end_time = datetime.now()
        time_taken = end_time - start_time
        time_taken = time_taken.microseconds
        
        times.append(time_taken)
    return times
        
```

```python
fibslow_time = time_function(fibslow , from_n , to_n)
fibfast_time = time_function(fibfast , from_n , to_n)

print(fibslow_time)
print(fibfast_time)
```

```python
plt.figure()
plt.plot(x , fibslow_time , label = "Fib Slow")
plt.plot(x , fibfast_time , label = "Fib Fast")
plt.xlabel("n")
plt.ylabel("Time (microseconds)")
plt.legend()
```
